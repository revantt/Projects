```python
#importing dependencies
import cv2 
import matplotlib.pyplot as plt
import os
```

## Algorithm
The algorithm I used does the following:-
<br>1.Reads the image 
<br>2.Converts the image into gray scale
<br>3.Checks every pixel on the left edge and the right edge for any pixel that is not black or any shade of black
<br>4.Notes down the pixel values of top and bottom of the strip
<br>5.Carves out the useful part of the image using the calculated coordinates
<br>6.Display the result

```python

def CropAlgo(path):
    import matplotlib.pyplot as plt
    img=cv2.imread(path) #Reading the image
    print("ORIGINAL IMAGE for",path,)
    plt.axis("off")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
    img= cv2.resize(img,(300,300))   #Resizing the image
    gr=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)   #converting the image to grayscale
    st=0
    row,col=gr.shape
    for i in range(row):   #Calculating the edges of the strip
        if((gr[i][0]>150 or gr[i][col-1]>150) and st==0):  #This determines the topmost coordinate of the strip
            st=1
            r1=i
        elif((gr[i][col-1]>150 or gr[i][0]>150) and st==1):  #This determines the bottom coordinate of the strip
            r2=i
    roi=img[r1-1:r2+1,0:col,:]  #The region of image i.e the bacteria strip plus 1 pixel of offset
    print("OUTPUT for",path)
    plt.axis("off")
    plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
    plt.show()
CropAlgo('IMG_3065.jpg')
```
